```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

def time_line(Lake):
    # Supposons que 'data' est votre DataFrame
    data_Lac = data[data['Lakes_names'] == Lake]
    data_Lac['Date'] = pd.to_datetime(data_Lac['Date'])

    df_sorted = data_Lac.sort_values(by='Date')

    # Mapper les catÃ©gories Ã  des valeurs numÃ©riques pour 'Depth'
    depth_mapping = {'E': 0, 'H': 1, 'M': 2}
    df_sorted['Depth_Num'] = df_sorted['Depth'].map(depth_mapping)

    # DÃ©finir une condition pour les triangles
    triangle_condition = df_sorted['Fraction_Size'] == 'L'

    plt.figure(figsize=(10, 6))  # DÃ©finir la taille de la figure

    # Tracer les points avec la couleur basÃ©e sur 'Depth_Num'
    scatter = plt.scatter(df_sorted['Date'], df_sorted[' contigs (>= 10000 bp)'], c=df_sorted['Depth_Num'], cmap='viridis', s=30, label='S')

    # Tracer les triangles pour les points qui satisfont la condition
    plt.scatter(df_sorted[triangle_condition]['Date'], df_sorted[triangle_condition][' contigs (>= 10000 bp)'], c=df_sorted[triangle_condition]['Depth_Num'], cmap='viridis', s=100, label='L')

    # Ajouter une barre de couleur avec l'Ã©tiquette 'Depth'
    plt.colorbar(scatter, ticks=np.arange(len(depth_mapping)), label='Depth', format=plt.FuncFormatter(lambda val, pos: list(depth_mapping.keys())[list(depth_mapping.values()).index(val)]))

    plt.title('Number of contigs >= 10 000 pb through the time for the lake : ' + str(Lake))  # Titre du graphique
    plt.xlabel('Date')  # Label de l'axe x
    plt.ylabel('Nombre de contigs >= 10 000 pb')  # Label de l'axe y
    plt.xticks(rotation=45)  # Rotation des Ã©tiquettes de l'axe x pour une meilleure lisibilitÃ©
    plt.grid(True)  # Afficher une grille

    plt.legend()  # Afficher la lÃ©gende
    plt.tight_layout()  # Ajuster automatiquement la disposition pour Ã©viter que les Ã©tiquettes ne se chevauchent
    plt.show()

#for i in range(0,len(list(data['Lakes_names'].unique()))):
#    time_line(list(data['Lakes_names'].unique())[i])

time_line("RIMV")
```
