```python
data = pd.read_csv("Plastid tree.csv", sep =";", header= 0)
data = data.fillna(0)
data.head()
#data.columns
```

```python
id_list1 = data['ID'].tolist()
id_list1 = [x for x in id_list1 if not isinstance(x, int)]
#print(id_list)

ref_list = data['RbcL_ID'].tolist()
#print(len(ref_list))
ref_list=listUnique(ref_list)
print("ref : ",ref_list)

Taxa_list = data['RbcL_based_taxa'].tolist()
print("Taxa : ",Taxa_list)

def listUnique(liste):
    L = []
    for i in range(0, len(liste)):
        if liste[i] in L:
            pass
        else:
            L.append(liste[i])
    return L

def transformer_liste(liste):
    nouvelle_liste = []
    for chaine in liste:
        if "DREP" in chaine:
            nouvelle_liste.append(chaine)
        else:
            # Appliquer la transformation en supprimant après le dernier underscore
            if '_' in chaine:
                resultat = chaine.rsplit('_', 1)[0]  # Utiliser rsplit pour diviser après le dernier underscore
            else:
                resultat = chaine  # Si aucun underscore, garder la chaîne originale
            nouvelle_liste.append(resultat)
    
    # Utiliser un set pour obtenir les éléments uniques et retourner une liste
    L = listUnique(nouvelle_liste)
    return L

id_list=transformer_liste(id_list1)
print("data : ", id_list)
```

```python
import tkinter as tk
from tkinter import ttk

def extract_before_last_underscore(text):
    if '_' in text:
        return '_'.join(text.rsplit('_', 1)[:-1])
    return text

# Prétraiter df pour conserver seulement la partie avant le dernier underscore
df['ID'] = df['ID'].apply(extract_before_last_underscore)

# Classe pour afficher et gérer les données
class DataViewer:
    def __init__(self, root):
        self.i = 0  # Initial index
        self.total = len(Final_ID)  # Total number of rows
        self.root = root
        self.root.title("Data Viewer")

        # Label to display information
        self.info_label = tk.Label(self.root, text="", font=("Arial", 14))
        self.info_label.pack(pady=10)

        # Treeview to display dataframe
        self.tree = ttk.Treeview(self.root, columns=('ID', 'Taxa'), show='headings')
        self.tree.heading('ID', text='ID')
        self.tree.heading('Taxa', text='Taxa')
        self.tree.pack(pady=10)

        # Buttons frame
        self.button_frame = tk.Frame(self.root)
        self.button_frame.pack(pady=10)

        # Previous button
        self.prev_button = tk.Button(self.button_frame, text="Previous", command=self.show_previous)
        self.prev_button.grid(row=0, column=0, padx=10)

        # Next button
        self.next_button = tk.Button(self.button_frame, text="Next", command=self.show_next)
        self.next_button.grid(row=0, column=1, padx=10)

        # Label for page number
        self.page_label = tk.Label(self.root, text="")
        self.page_label.pack(pady=10)

        # Show the first row of data
        self.show_next()

    def update_page_label(self):
        # Update the page label with current page number
        self.page_label.config(text=f"Page {self.i + 1} of {self.total}")

    def show_data(self):
        # Display the current data in the label and treeview
        ligne = data.iloc[self.i]
        rbcL_id_value = ligne['RbcL_ID']
        rbcL_based_taxa = ligne['RbcL_based_taxa']

        # Update label
        self.info_label.config(text=f"{rbcL_id_value}, {rbcL_based_taxa}")

        # Clear the Treeview
        self.tree.delete(*self.tree.get_children())

        # Filter dataframe df based on the RbcL_ID value
        filtered_df = df[df['ID'] == rbcL_id_value]

        # Insert filtered data into Treeview
        for index, row in filtered_df.iterrows():
            self.tree.insert('', 'end', values=(row['ID'], row['Taxa']))

        # Update the page label
        self.update_page_label()

    def show_next(self):
        # Increment index and wrap around if needed
        if self.i < len(Final_ID) - 1:
            self.i += 1
        else:
            self.i = 0  # Loop back to the first row

        self.show_data()

    def show_previous(self):
        # Decrement index and wrap around if needed
        if self.i > 0:
            self.i -= 1
        else:
            self.i = len(Final_ID) - 1  # Loop back to the last row

        self.show_data()

# Create the Tkinter window
root = tk.Tk()
app = DataViewer(root)

# Start the Tkinter event loop
root.mainloop()
```
